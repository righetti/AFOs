```python
oscill.initialize(K,omegaF,lamb)
init = np.array([0,10.])
oscill.integrate(0.,t_end,init,dt,save_dt)
```

```python
t = oscill.t()
phi = oscill.y()[0,:]
omega = oscill.y()[1,:]
plt.subplot(2,1,1)
plt.plot(t, phi)
plt.subplot(2,1,2)
plt.plot(t, omega)

om = omega[10000:]
print( (lamb*np.min(om) - omegaF)/lamb)
print( (lamb*np.max(om) - omegaF)/lamb)
print((lamb*np.mean(om) - omegaF)/lamb)
```
