```python
%matplotlib notebook
import numpy as np
import matplotlib.pylab as plt

import importlib

import pyafos
```

```python
K =  10000.
omegaF = 30
lamb = 1.
dt = 0.00001
save_dt = 0.001
t_end = 20.

oscill = pyafos.PhaseAFO()
```

```python
oscill.initialize_sine(K,omegaF,lamb)
init = np.array([0,10.])
oscill.integrate(0.,t_end,init,dt,save_dt)
```

```python
t = oscill.t()
phi = oscill.y()[0,:]
omega = oscill.y()[1,:]
plt.figure()
plt.subplot(2,1,1)
plt.plot(t, phi)
plt.subplot(2,1,2)
plt.plot(t, omega)

om = omega[10000:]
print( (lamb*np.min(om) - omegaF)/lamb)
print( (lamb*np.max(om) - omegaF)/lamb)
print((lamb*np.mean(om) - omegaF)/lamb)
```

```python
# test with time varying freq
omegaC = 1.
omegaF = 10.
oscill.initialize_frequency_changing_sine(K,omegaF,omegaC,lamb)
init = np.array([0,10.])
oscill.integrate(0.,t_end,init,dt,save_dt)
```
